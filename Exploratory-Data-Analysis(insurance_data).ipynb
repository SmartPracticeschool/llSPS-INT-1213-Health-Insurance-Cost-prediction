```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#importing dataset
insurance = pd.read_csv("insurance.csv")
insurance
```

```python
def sex(x):
    if x=='female':
        return 1
    else:
        return 0
insurance['sex'] = insurance['sex'].apply(sex)
insurance
```

```python
def smoker(x):
    if x=='yes':
        return 1
    else:
        return 0
insurance['smoker'] = insurance['smoker'].apply(smoker)
insurance
```

```python
def region(x):
    if x=='southeast':
        return 1
    elif x=='southwest':
        return 2
    elif x=='northeast':
        return 3
    else:
        return 4
insurance['region'] = insurance['region'].apply(region)
insurance
```
